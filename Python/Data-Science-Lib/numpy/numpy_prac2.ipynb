#Build Linear Regression FROM SCRATCH using Gradient Descent

```python
import numpy as np
```

```python
X=np.linspace(0,10,100)
y=3*X+7+np.random.randn(100)*2
```

```python
X=X.reshape(-1,1)
y=y.reshape(-1,1)
```

```python
m=np.zeros((1,1))
b=np.zeros((1,1))
lr=0.001
epochs=1000
```

```python
for i in range(epochs):
    y_pred = X.dot(m) + b
    
    error = y_pred - y
    
    m_grad = (1/len(X)) * np.dot(X.T, error)
    b_grad = (1/len(X)) * np.sum(error)
    

    m -= lr * m_grad
    b -= lr * b_grad
    
    if i % 100 == 0:
        loss = np.mean(error**2)
        print(f"Epoch {i}: Loss = {loss}")
```

```python
print("Slope (m):", m)
print("Intercept (b):", b)
```
