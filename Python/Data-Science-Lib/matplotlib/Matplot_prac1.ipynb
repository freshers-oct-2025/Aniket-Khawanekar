```python
from matplotlib import pylab
```

```python
import numpy as np


x=np.linspace(0,10,25)
y=x*x+2
print(x)
print(y)
# print(np.array([x,y]).reshape(25,2))
```

```python
pylab.plot(x,y,'r')
```

```python
pylab.subplot(2,2,1)
pylab.plot(x,y,'r--')
pylab.subplot(2,2,1)
pylab.plot(y,x,'g*-')
```

```python
from matplotlib import pyplot as plt
```

```python
fig=plt.figure()
axis=fig.add_axes([0.5,0.1,0.8,0.8])
axis.plot(x,y,'r')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
for ax in axes:
    ax.plot(x,y,'r')
```

```python
fig=plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.6,0.2,0.2])

axes1.plot(x,y,'r')
axes2.plot(x,y,'g')
```

```python
fig=plt.figure(figsize=(16,9),dpi=300)
axis=fig.add_axes([0.5,0.1,0.8,0.8])
axis.plot(x,y,'r')
```
